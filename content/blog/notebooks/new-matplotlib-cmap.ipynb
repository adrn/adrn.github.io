```python
from PIL import Image
import requests
from io import BytesIO

import numpy as np
from scipy.stats import multivariate_normal
import matplotlib
import matplotlib.pyplot as pl
pl.style.use('apw-notebook')
%matplotlib inline
```

```python
xy = multivariate_normal.rvs(mean=[0,0], size=10000)
z = multivariate_normal.pdf(xy, mean=[0,0], cov=[[1.,0],[0,1]])
pt_data = xy.T
```

```python
im_url = "https://pbs.twimg.com/profile_images/616542814319415296/McCTpH_E.jpg"
response = requests.get(im_url)
img = Image.open(BytesIO(response.content))
im_data = np.array(img)
```

```python
cmap = 'jet'

fig,axes = pl.subplots(1,2,figsize=(10,5))

axes[0].scatter(pt_data[0], pt_data[1], c=z, cmap=cmap)
axes[0].set_xlim(-2,2)
axes[0].set_ylim(-2,2)

axes[1].imshow(im_data[...,0], cmap=cmap)
```
